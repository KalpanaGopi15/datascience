# 7)Test the Analysis of Variance between etest_p and mba_p at signifance
level 5%.(Make decision using Hypothesis Testing)

We have to do analysis of variance between enterence test pass mark and mba pass mark. it is one way classification, we do analysis in column wise.

we get stats and p value. 
if P_value <0.05 then we can reject the Null Hypothesis. 

H0=Null hypothesis = there is no significant variance between enterence test pass mark and mba pass mark.
H1 = Alternative Hypothesis= There is significant difference between enterence test pass mark and mba pass mark.

Result:  we got the P_Value as 4.6725e-21 which is smaller than 0.05 so we can reject null hypothesis.

we conculde that there is significant difference between mba pass mark and enterence pass mark.



```python
import scipy.stats as stats
```

```python
stats.f_oneway(dataset['etest_p'],dataset['mba_p'])
```

# 8)Test the similarity between the degree_t(Sci&Tech) and specialisation(Mkt&HR) with respect to salary at significance level of 5%.(Make
decision using Hypothesis Testing)

h0=Null Hypothesis: There is no significance difference between degree_t(sci& Tech) with salary 
H1= Alternative hypothesis: There is significance difference between degree_t(sci& Tech) with salary
H0=Null Hypothesis: There is no significance differecne between Specialisation(MKT&HR) with salary 
H1= Alternative hypothesis: There is significance difference between Specialisation(MKT&HR) with salary 


```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
import scipy.stats as stats
model = ols('salary ~ C(degree_t) + C(specialisation)', data=placed).fit()
```

```python
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```

```python
model.summary()
```

```python
scitech=model.pvalues['C(degree_t)[T.Sci&Tech]']
mk=model.pvalues['C(specialisation)[T.Mkt&HR]']
print("sci&tech pvalue is:",scitech)
print("Market & HR p_value is:",mk)
```

both P_values are less than 0.05, so null hypotheses has been rejected. H1 alternative hypothesis accepted.

# 9)Convert the normal distribution to standard normal distribution for salary column

we have to write function to convert normal distribution to standard normal distribution for salary. we have to find mean and standard deviation for salary column. pass the dataset salary column into values by usinf for loop.To determine z_score/stdnormal distribution j as value -mean/standard deviation.

```python
def stdndis(dataset):
    import seaborn as sns
    mean=dataset.mean()
    std=dataset.std()
    values=[i for i in dataset]
    z_score=[((j-mean)/std) for j in values]
    sns.distplot(z_score,kde=True)
    sum(z_score)/len(z_score)
    print("Mean", mean, "std",std)
    
```

```python
stdndis(dataset['salary'])
```
